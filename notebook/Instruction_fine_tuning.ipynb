# Section 3: Utility Functions for PDF Processing & Data Handling

```python
def extract_pdf_text(pdf_path: str) -> List[Dict[str, Any]]:
    """
    Extract text from PDF file with page information

    Args:
        pdf_path: Path to PDF file

    Returns:
        List of dictionaries containing page number and text
    """
    pages = []

    try:
        with fitz.open(pdf_path) as doc:
            for page_num, page in enumerate(doc, start=1):
                text = page.get_text("text").strip()
                if text:
                    pages.append({
                        "page": page_num,
                        "text": text,
                    })
    except Exception as e:
        print(f"Error extracting PDF: {e}")
        return []

    print(f"✓ Extracted {len(pages)} pages from PDF")
    return pages


def clean_text(text: str) -> str:
    """
    Clean and normalize text for training

    Args:
        text: Raw text to clean

    Returns:
        Cleaned text
    """
    # Unicode normalization
    text = unicodedata.normalize("NFKC", text)
    text = text.replace("\u200b", "").replace("\ufeff", "")

    # Fix hyphenated words broken across lines
    text = re.sub(r"(\w)-\n(\w)", r"\1\2", text)

    # Remove page numbers
    text = re.sub(r"(?m)^\s*\d+\s*$", "", text)

    # Normalize spaces
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text)

    return text.strip()


def chunk_text(text: str, chunk_size: int = 512, overlap: int = 50) -> List[str]:
    """
    Split text into overlapping chunks

    Args:
        text: Text to chunk
        chunk_size: Maximum characters per chunk
        overlap: Number of overlapping characters between chunks

    Returns:
        List of text chunks
    """
    chunks = []
    start = 0

    while start < len(text):
        end = min(start + chunk_size, len(text))

        # Try to break at sentence boundary
        if end < len(text):
            last_period = text.rfind(".", start, end)
            if last_period > start:
                end = last_period + 1

        chunk = text[start:end].strip()
        if len(chunk) >= MIN_CHARS_PER_CHUNK:
            chunks.append(chunk)

        start = end - overlap if end < len(text) else end

    return chunks


def clear_gpu_memory():
    """Clear GPU memory cache"""
    gc.collect()
    torch.cuda.empty_cache()


def print_trainable_params(model):
    """Print model trainable vs total parameters"""
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    total_params = sum(p.numel() for p in model.parameters())

    print(f"Trainable params: {trainable_params:,} / {total_params:,}")
    print(f"Trainable %: {100 * trainable_params / total_params:.2f}%")

print("✓ Utility functions defined")
```

```python
def create_dataset_from_pdf(pdf_path: str) -> Dataset:
    """
    Create training dataset from PDF file

    Args:
        pdf_path: Path to PDF file

    Returns:
        HuggingFace Dataset
    """
    pages = extract_pdf_text(pdf_path)
    records = []

    for page in pages:
        cleaned_text = clean_text(page["text"])
        chunks = chunk_text(cleaned_text, chunk_size=512, overlap=CHUNK_OVERLAP)

        for chunk_id, chunk in enumerate(chunks):
            if len(chunk) >= MIN_CHARS_PER_CHUNK:
                records.append({
                    "text": chunk,
                    "source_page": page["page"],
                    "chunk_id": chunk_id,
                })

    print(f"✓ Created {len(records)} text chunks from PDF")
    if records:
        print(f"Sample chunk:\n{records[0]['text'][:300]}...")

    return Dataset.from_list(records)


def load_instruction_dataset_from_jsonl(jsonl_path: str) -> Dataset:
    """
    Load instruction dataset from JSONL file

    Args:
        jsonl_path: Path to JSONL file with format:
                   {"instruction": "...", "input": "...", "output": "..."}

    Returns:
        HuggingFace Dataset
    """
    records = []

    try:
        with open(jsonl_path, 'r') as f:
            for line in f:
                if line.strip():
                    record = json.loads(line)
                    # Ensure required fields exist
                    record.setdefault('instruction', '')
                    record.setdefault('input', '')
                    record.setdefault('output', '')
                    records.append(record)
    except Exception as e:
        print(f"Error loading JSONL: {e}")
        return Dataset.from_list([])

    print(f"✓ Loaded {len(records)} instruction examples from JSONL")
    if records:
        print(f"Sample instruction:\n{records[0]}")

    return Dataset.from_list(records)


def format_instruction_prompt(instruction: str, input_text: str = "", output_text: str = "") -> str:
    """
    Format instruction in alpaca-style template

    Args:
        instruction: Main instruction
        input_text: Optional input context
        output_text: Expected output (for training)

    Returns:
        Formatted prompt string
    """
    instruction = str(instruction).strip()
    input_text = str(input_text).strip()
    output_text = str(output_text).strip()

    if input_text:
        prompt = f"### Instruction:\n{instruction}\n\n### Input:\n{input_text}\n\n### Response:\n{output_text}"
    else:
        prompt = f"### Instruction:\n{instruction}\n\n### Response:\n{output_text}"

    return prompt


def format_instruction_dataset(dataset: Dataset) -> Dataset:
    """
    Format instruction dataset with proper prompts

    Args:
        dataset: Raw instruction dataset

    Returns:
        Formatted dataset with 'text' field
    """
    def format_row(row):
        prompt = format_instruction_prompt(
            instruction=row.get('instruction', ''),
            input_text=row.get('input', ''),
            output_text=row.get('output', '')
        )
        return {"text": prompt}

    formatted = dataset.map(format_row, remove_columns=dataset.column_names)
    print(f"✓ Formatted {len(formatted)} instruction examples")
    if len(formatted) > 0:
        print(f"Sample formatted text:\n{formatted[0]['text'][:300]}...")

    return formatted

print("✓ PDF and instruction processing functions defined")
```

# Section 4: Load Tokenizer and Model with LoRA/QLoRA

```python
def load_model_with_lora(model_name_or_path: str, load_in_4bit: bool = True):
    """
    Load model with LoRA/QLoRA configuration using Unsloth

    Args:
        model_name_or_path: Model name or path
        load_in_4bit: Whether to use 4-bit quantization (QLoRA)

    Returns:
        Tuple of (model, tokenizer)
    """
    print(f"\nLoading model: {model_name_or_path}")
    print(f"4-bit Quantization (QLoRA): {load_in_4bit}")

    # Load model and tokenizer using Unsloth
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=model_name_or_path,
        max_seq_length=MAX_SEQ_LENGTH,
        dtype=None,
        load_in_4bit=load_in_4bit,
    )

    # Setup tokenizer
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "right"

    # Apply LoRA configuration
    print("\nApplying LoRA configuration:")
    print(f"  LoRA Rank (r): {LORA_R}")
    print(f"  LoRA Alpha: {LORA_ALPHA}")
    print(f"  LoRA Dropout: {LORA_DROPOUT}")

    model = FastLanguageModel.get_peft_model(
        model,
        r=LORA_R,
        lora_alpha=LORA_ALPHA,
        target_modules=LORA_TARGET_MODULES,
        lora_dropout=LORA_DROPOUT,
        bias="none",
        use_gradient_checkpointing="unsloth",
        random_state=SEED,
    )

    print("\nModel loaded successfully!")
    print_trainable_params(model)

    return model, tokenizer


# Load base model
print("=" * 60)
print("LOADING BASE MODEL")
print("=" * 60)

base_model, base_tokenizer = load_model_with_lora(BASE_MODEL, load_in_4bit=True)
clear_gpu_memory()

print("\n✓ Base model ready for training")
```

# Section 9: Inference Functions and Interactive Bot

```python
def generate_response(model, tokenizer, instruction: str, input_text: str = "", max_new_tokens: int = 256) -> str:
    """
    Generate response using the model

    Args:
        model: Fine-tuned model
        tokenizer: Model tokenizer
        instruction: User instruction
        input_text: Optional input context
        max_new_tokens: Maximum tokens to generate

    Returns:
        Generated response
    """
    # Prepare prompt
    FastLanguageModel.for_inference(model)

    prompt = format_instruction_prompt(instruction, input_text)
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=MAX_SEQ_LENGTH).to("cuda")

    # Generate
    with torch.inference_mode():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            repetition_penalty=1.1,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )

    # Decode response
    response = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True).strip()
    return response


def interactive_bot(model, tokenizer, stage_name: str):
    """
    Interactive bot for testing model

    Args:
        model: Fine-tuned model
        tokenizer: Model tokenizer
        stage_name: Name of the model stage
    """
    print(f"\n{'='*60}")
    print(f"Interactive Bot - {stage_name}")
    print(f"{'='*60}")
    print("Enter 'quit' to exit\n")

    test_queries = [
        ("What causes CT Scanner failure scenario?", ""),
        ("What is the function of the Mammography component", ""),
        ("Why is image quality degraded on MRI case?", ""),
    ]

    for instruction, input_text in test_queries:
        print(f"\n[User] Instruction: {instruction}")
        if input_text:
            print(f"[User] Input: {input_text}")

        response = generate_response(model, tokenizer, instruction, input_text)
        print(f"[{stage_name}] Response: {response[:500]}...")
        print("-" * 60)


def batch_generate_responses(model, tokenizer, test_queries: List[Tuple[str, str]]) -> List[str]:
    """
    Generate responses for multiple queries

    Args:
        model: Fine-tuned model
        tokenizer: Model tokenizer
        test_queries: List of (instruction, input) tuples

    Returns:
        List of generated responses
    """
    responses = []
    for instruction, input_text in test_queries:
        response = generate_response(model, tokenizer, instruction, input_text)
        responses.append(response)

    return responses

print("✓ Inference functions defined")
```
